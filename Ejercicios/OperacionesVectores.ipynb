```python
import numpy as np
import scipy.spatial as dis

q = np.random.randint(20, size=(1, 4))                      
entrenamiento = np.random.randint(20, size=(10, 4))
#Ejemplo 1
distancia = []
print(f"Datos de query:\n{q}\n\nDatos de entrenamiento:\n{entrenamiento}\n")

for i in range(entrenamiento.shape[0]):
    z_q_entr_i = q[0,:] - entrenamiento[i,:]
    dista = (z_q_entr_i @ z_q_entr_i.T) ** (1/2) # -> (1, 4) * (4, 1) = (1,1) -> (1, 1) ** .5 
    distancia.append(dista)

print("Distancias ejercicio 1:")
for d in distancia:
    print(d)
print()
#Ejemplo 2
Q = np.tile(q,(entrenamiento.shape[0],1))
print(f'Matriz Q:\n{Q}\n')
Z = Q - entrenamiento
print(f'Matriz Z:\n{Z}\n')
S_with_garbage = Z @ Z.T
print(f'Matriz con basura:\n{S_with_garbage}\n')
distancias_S = np.diag(S_with_garbage) ** (1/2)

print("Distancias ejercicio 2:")
for d in distancias_S:
    print(d)
```
