```python
i = 6000      
freqs = []
while i >= 5:
    freqs.append(i)
    i //= 2
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k','w', '#FFA07A', '#6A5ACD', '#00FF00', '#FFD700', '#8B4513']
freqs
```

```python
for f in freqs:
    dict_perc = {}
    for p in range(20, 61, 5):
        for k in [3, 5]:
            ruta = f"Seleccion/PruebaPorcentajef{f}/porcentajeExitok{k}p{p}f{f}.csv"
            df_perEx = pd.read_csv(ruta)
            dict_perc[str((k,p))] = df_perEx["Porcentaje"].mean()
    file = f"Seleccion/PruebasMeans/pruebaf{f}.json"
    with open(file, 'w') as file_save:
        json.dump(dict_perc,file_save)
```

```python
plt.figure(figsize=(15,7))
plt.title('Porcentajes promedios de 1000 experimentos individuales',size=20)

for f,c in zip(freqs, colors):
    with open(f"Seleccion/PruebasMeans/pruebaf{f}.json") as f_ile:
        file = json.load(f_ile)
    xcoor = [key for key in file.keys()]
    ycoor = [values for values in file.values()]
    plt.plot(xcoor, ycoor, c, label=f'Frecuencia {f}')


plt.xlabel('(k, perc)',size=20)
plt.ylabel('Porcentaje %',size=20)
plt.ylim(50,100)
plt.xticks(rotation=90)  # Rotar las etiquetas del eje x para mayor claridad si son largas
plt.tight_layout()
plt.grid(True)
plt.legend()
plt.show()
```
